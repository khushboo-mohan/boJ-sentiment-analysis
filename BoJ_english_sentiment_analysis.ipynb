```python
SHEET_ID = "12JfNLn-dtRj3IVZLVCCYYurSia10IAAsDuMwBAlPFPQ"  # Replace with your actual Sheet ID
SHEET_NAME = "Sheet1"       # Update if your sheet has a different name

# Open Google Sheet
sh = gc.open_by_key(SHEET_ID)
worksheet = sh.worksheet(SHEET_NAME)

# Convert to Pandas DataFrame
import pandas as pd
data = pd.DataFrame(worksheet.get_all_records())

# Show first few rows
print(data.head())
```

```python
from torch.nn.functional import softmax

def analyze_sentiment(text):
    # Tokenize the input speech
    inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=512)

    # Run model inference
    with torch.no_grad():
        outputs = model(**inputs)

    # Convert logits to probabilities
    probs = softmax(outputs.logits, dim=1)

    # Define sentiment categories
    labels = ["Negative", "Neutral", "Positive"]

    # Get the highest probability sentiment
    sentiment = labels[torch.argmax(probs)]

    # Return sentiment label and confidence score
    return sentiment, probs.max().item()
```

```python
# Apply sentiment analysis to each speech
data["Sentiment"], data["Confidence"] = zip(*data["Speech"].apply(analyze_sentiment))

# Show updated dataframe
print(data.head())
```

```python
# Convert DataFrame back to list
data_list = [data.columns.values.tolist()] + data.values.tolist()

# Update Google Sheets
worksheet.update(data_list)
print("✅ Sentiment scores added to Google Sheets!")
```

```python
import paddlehub as hub
senta = hub.Module(name="senta_bilstm")  # Load the model
```

```python
# Step 1: Import required libraries
from google.colab import auth
import gspread
from google.auth import default

# Step 2: Authenticate user
auth.authenticate_user()

# Step 3: Authorize with Google Sheets
creds, _ = default()
gc = gspread.authorize(creds)

# Step 4: Open the Google Sheet (Replace "BOJ_Speeches" with your actual sheet name)
sheet = gc.open("BOJ_Speeches")
worksheet = sheet.sheet1  # Use `sheet.worksheet("YourSheetName")` if needed

# Step 5: Read text from column G (row 2 onwards, assuming row 1 is a header)
texts = worksheet.col_values(7)[1:]  # Column G is index 7, skipping header row

# Step 6: Perform sentiment analysis (assuming `senta` is already loaded)
if texts:
    results = senta.sentiment_classify(texts=texts)  # Ensure `senta` is preloaded

    # Extract sentiment labels (e.g., Positive/Negative)
    sentiment_labels = [res["sentiment_label"] for res in results]

    # Step 7: Write results back to column H (row 2 onwards)
    for i, sentiment in enumerate(sentiment_labels, start=2):
        worksheet.update_cell(i, 8, sentiment)  # Column H (Index 8)

print("✅ Sentiment analysis completed and results updated in Google Sheets!")
```
